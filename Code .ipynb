# Code For Data Cleaning.

```python
# Importing various libraries as per requirment. 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Converting the csv file to a dataframe.

df=pd.read_csv("Unclean movies list.csv")
```

```python
# Checking the Dataframe.

df
```

```python
# Dropping the GROSS column 

df=df.drop(["Gross"],axis=1)
```

```python
df
```

```python
# Dropping the Duplicate values in the movies section.

df=df.drop_duplicates(["MOVIES"])
```

```python
df
```

```python
# Dropping the ONE-LINE, STARS, RunTime

df=df.drop(["ONE-LINE","STARS","RunTime"],axis=1)
```

```python
df
```

```python
# Dropping the Null values 

df=df.dropna()
```

```python
df
```

```python
# Putting the condition that only those films of Rating more than 9 are to be selected.

df=df.loc[(df[:]["RATING"])>9.0]
```

```python
# Resetting the indexes.

df=df.reset_index(drop=True)
```

```python
# Removing all the "\n" from the contents.

df["GENRE"]=df["GENRE"].str.lstrip("\n")
```

```python
df
```

```python
df
```

```python
# Removing the comma so that no prblm comes.

df["VOTES"]=df["VOTES"].str.replace(",","")
```

```python
# Converting the votes type to integer from string.

df["VOTES"]=df["VOTES"].astype("int")
```

```python
# Checking the info of the dataframe.

df.info()
```

```python
df
```

```python
# Sorting the values of votes on the basis of higher votes so that decision can be taken.

df=df.sort_values("VOTES",ascending=False)
```

```python
df.drop()
```

```python
# Taking the only first 5 movies for further decision making.

df=df.head(3)
```

```python
df
```

```python
df.to_csv("Clean Movie Data.csv")
```
